```python
import yfinance as yf
import pandas as pd
import empyrical as ep
import pyfolio as pf

# Define the stock symbol and date range
stock_symbol = 'AAPL'  # Replace with your desired stock symbol
start_date = '2020-01-01'
end_date = '2023-01-01'

# Fetch historical price data using yfinance
data = yf.download(stock_symbol, start=start_date, end=end_date)

# Create moving averages and signals
data['MA50'] = data['Close'].rolling(window=50).mean()
data['MA200'] = data['Close'].rolling(window=200).mean()

data['Signal'] = 0
data.loc[data['MA50'] > data['MA200'], 'Signal'] = 1

# Convert signals into returns
data['Returns'] = data['Signal'].shift(1) * data['Close'].pct_change()

# Calculate strategy performance metrics
returns = data['Returns']
strategy_returns = returns.dropna()
strategy_cumulative_returns = ep.cum_returns(strategy_returns, starting_value=1.0)
strategy_annual_returns = ep.annual_return(strategy_returns)
strategy_volatility = ep.annual_volatility(strategy_returns)
strategy_sharpe_ratio = ep.sharpe_ratio(strategy_returns)

# Create PyFolio tear sheet
pf.create_simple_tear_sheet(strategy_returns)
```
